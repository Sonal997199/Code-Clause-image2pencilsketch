```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread('/content/Dataset/vivek bhaiya.jpg')
```

```python
# convert the image to gray scale
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
# invert the grey image
invert=cv2.bitwise_not(gray_image)
```

```python
# Blur the image by gaussian blur
blur=cv2.GaussianBlur(invert,(21,21),0)
```

```python
# Invert the blurred image
invertedblur=cv2.bitwise_not(blur)
```

```python
# Create the pencil sketch image
sketch_filter=cv2.divide(gray_image,invertedblur,scale=256.0)
```

```python
# Show the image
cv2_imshow(image)
```

```python
cv2_imshow(sketch_filter)
```
